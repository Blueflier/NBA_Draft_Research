```python
alreadyDone = []
for name in df1['PLAYER']:
        
    if name not in alreadyDone:
        
        try:
        
            #get overall position
            player = df[df['PLAYER_NAME']==name].head()
            overall = player['OVERALL_PICK'].values[0]

            #append overall position to name
            df1.loc[df1['PLAYER'] == name, 'overall'] = overall
            
            
        
        except:
            pass
        
        alreadyDone.append(name)
        
```

```python
draft_years_after = []

for row in df1.itertuples():
    

    
    try:
        player = df[df['PLAYER_NAME']==row.PLAYER].head()
        season = player['SEASON'].values[0]
        season = int(season)
        

        
        
        year = row.Year - season
        
    
        
        draft_years_after.append(year)
    except:
        draft_years_after.append(-1)

```
