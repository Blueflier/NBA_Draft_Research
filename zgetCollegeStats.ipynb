```python
def getPlayerStats(Id,league = 'nba'):
    from nba_api.stats.endpoints import playercareerstats
    import time
    time.sleep(3)
    career = playercareerstats.PlayerCareerStats(per_mode36 = 'PerGame',player_id=Id).season_totals_college_season.get_data_frame()
    return career
```

```python
import pandas as pd
def getCollegeData(ids):
    import numpy as np
    import pandas as pd
    import time
    if len(ids) > 1:
        initial = 0
        for ID in ids:
            if initial == 0:
                current_df = getPlayerStats(ID,league ='college')
                initial = 1
            else:
                next_df = getPlayerStats(ID,league ='college').head(1)
                current_df = pd.concat([current_df, next_df], ignore_index=True)
                
            
    else:
        current_df = getPlayerStats(ID,league = 'college')
    
    data = current_df
        
    return data


physical_attributes = pd.read_csv('/Users/Tim/Documents/GitHub/NBA-Draft-Biola-Summer-Research/draft_data_2000_2016_no_empty_cells.csv')

#display(physical_attributes)

player_ids = physical_attributes["PLAYER_ID"].to_numpy()
df = getCollegeData(player_ids)
display(df)
df.to_csv('draft_pre_nba_data_2000_2016_no_empty_cells.csv', index=False)
```
