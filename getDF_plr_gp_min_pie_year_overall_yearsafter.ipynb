```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/josephhartono/Documents/GitHub/NBA-Draft-Biola-Summer-Research/PIE_overall_yearsAfter.csv')
df.reset_index()
# display(df[df['PLAYER'] == "Blake Griffin"])
display(df)
```

```python

for plr in set(df['PLAYER']):
    #print(plr)
    df1 = df[df['PLAYER'] == plr]
    #print(df1)
    
    latest_year = max(df1['yearsAfter'])
    years = df1['yearsAfter'].tolist()
    #display(years)
    for i in range(0,latest_year+1):
        if i not in years:
            df2 = df1.head(1)
            vals = df2.values[0]
            base_year = vals[5] - vals[7]
            df.loc[len(df.index)] = [-1,vals[1], 0, 0, 0, base_year + i, float(vals[6]), int(i)] 
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.to_csv('gp_min_pi_year_ovr_YA.csv')
```
