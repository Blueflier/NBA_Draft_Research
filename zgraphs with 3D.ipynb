```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
df = pd.read_csv('/Users/josephhartono/Documents/GitHub/NBA-Draft-Biola-Summer-Research/3D data.csv', index_col = [0])
df.rename(columns={"x":"overall"})
df.rename(columns={"y":"yearsAfter"})
# df.columns
df= df[df['y'] > -1]
df
```

```python
ax = plt.figure(dpi=300,figsize=(8,6)).add_subplot(projection='3d')
x = df['x']
y = df['norm_sal']



z = df['y']
ax.scatter(x, y, zs=z, zdir='y', s=42, c = (z<=3), cmap = 'coolwarm', label='First 4 Seasons')

ax.set_xlabel('Overall Draft Pick')
ax.set_ylabel('years After draft')
ax.set_zlabel('Normalized Salary')

ax.axes.get_yaxis().set_visible(False)
ax.axes.get_yaxis().set_ticks([0,10,20])

ax.axes.set_xlim3d(left=0.000001, right=61)
ax.axes.set_ylim3d(bottom=0, top=20)
ax.axes.set_zlim3d(bottom=0, top=1)

ax.view_init(elev=10., azim=-80)
plt.legend()
plt.show()
```
