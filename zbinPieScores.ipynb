```python



def salary_binning(filepath,start_year,end_year):
    import pandas as pd
    import numpy as np
    #salaries = pd.read_csv('/Users/Tim/Documents/GitHub/NBA-Draft-Biola-Summer-Research/nba_salaries_1984_to_2017.csv')
    #salaries = pd.read_csv('/Users/Tim/Documents/GitHub/NBA-Draft-Biola-Summer-Research/nba_salaries_2018_to_2021.csv')
    salaries = pd.read_csv(filepath)
    #years = salaries['season_end'].values
    #years = np.arange(2019,2023, 1)1985,2018
    years = np.arange(start_year,end_year,1)
    
    #print(salaries['salary'])
    initial = 0
    for yr in years:
        if initial  == 0:
            df_current = salaries[salaries['season_end'] == yr].copy()
            df_current['salary_bin'] = pd.qcut(salaries['salary'],
                               q=[0, .2, .5, .8, .90, .95, 1],
                               labels=['minimum', 'below average', 'average', 'above average','near max', 'max'])
            initial = 1
        else:
            df_next = salaries[salaries['season_end'] == yr].copy()
            df_next['salary_bin'] = pd.qcut(salaries['salary'],
                               q=[0, .2, .5, .8, .90, .95, 1],
                               labels=['minimum', 'below average', 'average', 'above average','near max', 'max'])
            df_current = pd.concat([df_current, df_next], ignore_index=True)
    
    data_salaries = df_current
    return data_salaries
```
