```python
#data = pd.read_csv ('/Users/Tim/Documents/GitHub/NBA-Draft-Biola-Summer-Research/finished_joseph_csv_manual_pie - joseph_csv_manual_pie.csv')
data = pd.read_csv ('/Users/josephhartono/Downloads/finished_joseph_csv_manual_pie - joseph_csv_manual_pie.csv')
data.drop('PLAYER_ID', inplace=True, axis=1)
data.drop('FIRST_NAME_x', inplace=True, axis=1)
data.drop('LAST_NAME_x', inplace=True, axis=1)
data.drop('PLAYER_NAME_x', inplace=True, axis=1)
data.drop('HEIGHT_WO_SHOES_FT_IN', inplace=True, axis=1)
data.drop('WINGSPAN_FT_IN', inplace=True, axis=1)
data.drop('STANDING_REACH_FT_IN', inplace=True, axis=1)
data.drop('DRAFT_CLASS', inplace=True, axis=1)
data.drop('SEASON_ID', inplace=True, axis=1)
data.drop('GP', inplace=True, axis=1)
data.drop('GS', inplace=True, axis=1)
data.drop('MIN', inplace=True, axis=1)
data.drop('ORGANIZATION_ID', inplace=True, axis=1)
data.drop('SCHOOL_NAME', inplace=True, axis=1)
data.drop('PLAYER_AGE', inplace=True, axis=1)
data.drop('LEAGUE_ID', inplace=True, axis=1)
#PG SG SF PF C
mapping = {'PG':1,'PG-SG':1,'PG-SF':1,'PG-PF':1,'PG-C':1,
           'SG':2,'SG-PG':2,'SG-SF':2,'SG-PF':2,'SG-C':2,
           'SF':3,'SF-PG':3,'SF-SG':3,'SF-PF':3,'SF-C':3,
           'PF':4,'PF-PG':4,'PF-SG':4,'PF-SF':4,'PF-C':4,
           'C':5,'C-PG':5,'C-SG':5,'C-SF':5,'C-PF':5}
data = data.replace({'POSITION_x':mapping})

#data['PIE_score_5YR'] = pd.qcut(data['PIE_score_5YR'], q=[0, .2, .4, .6, .8, 1], labels = [10 ,20, 30,40,50])
print(type(data["PIE_score_5YR"][0]))
#print(data.columns)
data
```

```python
#data

#splits the data
train_x = data.sample(frac = 0.7)
validation_x = data.sample(frac = 0.3)

#gets pie score of train_x
#gets pie score of validation_x
train_y = train_x["PIE_score_5YR"].values
train_y = np.asarray(train_y)

validation_y = validation_x["PIE_score_5YR"].values
validation_y = np.asarray(validation_y)


# display(train_y)
# display(validation_y)

#takes out pie score from train_x
#takes out pie score from validation_x
train_x.drop('PIE_score_5YR', inplace=True, axis=1)
validation_x.drop('PIE_score_5YR', inplace=True, axis=1)


#just gets the values of train_x without the headers
train_x = train_x.values
validation_x = validation_x.values

# train_y = np.asarray(train_y)
display(validation_y)




display(validation_y)

# validation_y = validation_y[:,None]
# train_y = train_y[:,None]
# display(validation_y)
# display(train_y)

# train_y = to_categorical(train_y, 5)
# validation_y = to_categorical(validation_y, 5)

# display(train_x)
# display(test_x)




```

```python
#BUILD THE MODEL
# define cnn model

model = keras.Sequential([
    #keras.layers.Conv2D(32, kernel_size = (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)),
    #keras.layers.MaxPooling2D( pool_size = (2, 2)),
    #keras.layers.Flatten(),
    keras.layers.Dense(139, activation='sigmoid'),
    keras.layers.Dense(100, activation='sigmoid'),
    keras.layers.Dense(29, activation='softmax')
])

# compile model
#opt = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer="adam", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
 model.fit(train_x, train_y, epochs = 10)
```

```python
test_loss, test_acc=model.evaluate(validation_x,validation_y)
print("Testing set loss :  %.3f"%test_loss)
#Eout.append((100-(100*test_acc)))
print("Testing set error :  %.2f"%(100-(100*test_acc)), "%")
```
