```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

df = pd.read_csv('nm_sal_yr_pos_gp_min_ovr_ya_nS.csv', index_col = [0])
grp = [5,10,15,30,60]
df
```

```python
df1 = df[df['overall'] == 1]
```

```python
df=df[df['yearsAfter'] == 4]
x_m = []
mean = []
std = []

for i in range(0,5):
    df1 = df[(df['overall'] >= grp[i]-(grp[i-1]-1)) & (df['overall'] <= grp[i])]#1 5 
    display(df1)
    break
    pie = df1['norm_sal'].values
    pie = np.array(pie)
    
    
    x_m.append(pie)
    mean.append(pie.mean())
    std.append(pie.std())
display(mean)
display(std)

mn = np.array(mean)
sd = np.array(std)

display(sd/mn)
```

```python
from matplotlib.pyplot import cm
plt.clf()

f1, ax2 = plt.subplots(dpi=300,figsize=(8,4))
color = ['tab:blue','tab:orange', 'tab:green','tab:red','tab:purple']

for i in range(0,5):

    a1, b1=np.histogram(x_m[i],20)
    xt=b1[1:] #get the rightmost coordinates
    dx=xt[1]-xt[0] #get the bin width
    x=xt-dx/2 #shift to the bin middle

    #to normalize the values of the bins so that we get a probability ditribution
    L=len(x_m[i]) #number of data points
    y=a1/L/dx #divide by bin size and total number of points


    c = color[i]
    labels = [f'1-{grp[0]}',f'{grp[0]+1}-{grp[1]}',f'{grp[1]+1}-{grp[2]}',f'{grp[2]+1}-{grp[3]}',f'{grp[3]+1}-{grp[4]}']
    ax2.plot(x,y,'-D',color=c, linewidth=2,markersize=4, label = labels[i]);
    
    
#=============
#matplotlib.pyplot.hlines(y, xmin, xmax, colors=None, linestyles='solid', label='', *, data=None, **kwargs)
# lw = 6
# al1 = 0.7
# al2 = 0.9
# ax2.vlines(x = mean[0]/(grp[0]), color = 'tab:blue',ymin = 0, ymax = std[0]/(grp[0]),linewidth=lw,alpha = al1)
# ax2.vlines(x = mean[1]/(grp[1]-grp[0]), color = 'tab:orange',ymin = 0, ymax = std[1]/(grp[1]-grp[0]),linewidth=lw,alpha = al1)
# ax2.vlines(x = mean[2]/(grp[2]-grp[1]), color = 'tab:green',ymin = 0, ymax = std[2]/(grp[2]-grp[1]),linewidth=lw,alpha = al1)
# ax2.vlines(x = mean[3]/(grp[3]-grp[2]), color = 'tab:red',ymin = 0, ymax = std[3]/(grp[3]-grp[2]),linewidth=lw,alpha = al2)
# ax2.vlines(x = mean[4]/(grp[4]-grp[3]), color = 'tab:purple',ymin = 0, ymax = std[4]/(grp[4]-grp[3]),linewidth=lw,alpha = al2)
sp = 0.5
h = 0.3
p = 6.5
for i in range(0,5):
    ax2.hlines(p - (sp * i), 0 + mean[i] - std[i]/2,std[i] + mean[i] - std[i]/2, color = color[i])
    ax2.vlines(x = mean[i], ymin = (p-h) - (sp * i), ymax = (p+h) - (sp * i) , color = color[i])

# ax2.set_xticks([-1, 0,1,2,3,4,5], ['-0.05','0', '.05', '.10','.15','.20','.25'])
# ax2.set_ylabel("Density Distribution")
# ax2.set_xlabel("Cumulative PIE score")
ax2.set_xlim(0, 0.8)
ax2.set_ylim(-.2, 7.0)
plt.legend(title="Overall Draft Pick")
ax2.tick_params(axis='x', labelsize=17)
ax2.tick_params(axis='y', labelsize=17)
#plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
plt.savefig('Year5Sal Hist.png')
plt.show()
```
