```python
import numpy as np
import pandas as pd
df = pd.read_csv('/Users/josephhartono/Downloads/nba_files/nnba_salaries_2018_to_2021_w_salary_tags.csv')
df2 = pd.read_csv('/Users/josephhartono/Downloads/nba_files/nba_salaries_1984_to_2017_w_salary_tags.csv')
df3 = pd.read_csv('/Users/josephhartono/Downloads/nba_files/nba2018-2022Positions.csv')
```

```python
position = []
count = 0
for name in df['name']:
    
    bug_name = name[:-1]
        
    try:
        copy = df2.copy(deep = True)
        copy = copy[copy['name'] == bug_name]
        copy = copy.iloc[0]
        position.append(copy['position'])
    except:
        copy = df3.copy(deep = True)
        copy = copy[copy['Player'] == bug_name]
        copy = copy.head(1)
        pos = copy['Pos'].values
        pos = pos[0]  
        position.append(pos)
df['position'] = position
```

```python
mapping = {'PG':"Point Guard",'PG-SG':"Point Guard and Shooting Guard",'PG-SF':"Point Guard and Small Forward",'PG-PF':"Point Guard and Power Forward",'PG-C':"Point Guard and Center",
           'SG':"Shooting Guard",'SG-PG':"Shooting Guard and Point Guard",'SG-SF':"Shooting Guard and Small Forward",'SG-PF':"Shooting Guard and Power Forard",'SG-C':"Shooting Guard and Center",
           'SF':"Small Forward",'SF-PG':"Small Forward and Point Guard",'SF-SG':"Small Forward and Shooting Guard",'SF-PF':"Small Forward and Power Forward",'SF-C':"Small Forward and Center",
           'PF':"Power Forward",'PF-PG':"Power Forward and Point Guard",'PF-SG':"Power Forward and Shooting Guard",'PF-SF':"Power Forward and Small Forward",'PF-C':"Power Forward and Center",
           'C':"Center",'C-PG':"Center and Point Guard",'C-SG':"Center and Shooting Guard",'C-SF':"Center and Small Forward",'C-PF':"Center and Power Forward"}
df = df.replace({'position':mapping})
```
